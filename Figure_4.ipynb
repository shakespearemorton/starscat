```python
k = 0
maxii = []
tippi = []
r = []
for i in data2:
    plt.plot(xn[3:],i[3:],'-.',label = y[k])
    maxii.append(np.max(i))
    tippi.append(np.argmax(i))
    r.append(np.max(i)-i[-1])
    k+=1
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
plt.ylabel('LSPR (nm)')
plt.xlabel('# of tips')
plt.savefig('hybridization2.png',type='png',dpi=1200,bbox_inches='tight')
```
